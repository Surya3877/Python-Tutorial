# **Python Comments**

# **Comments**

Python has commenting capability for the purpose of in-code documentation.

Comments start with a #, and Python will render the rest of the line as a comment:

**Example**

Comments in Python:

```python
#This is a comment.
print("Hello, World!")
```

Comments can be placed at the end of aline, and Python will ignore the rest of the line:

**Example**

```python
print("Hello, World!") #This is a comment
```

A comment does not have to be text that explains the code, It can be used to prevent Python from executing code:

**Example**

```python
#print("Hello, World!")
print("Hello, World!")
```

# **Multiline Comments**

Python does not realyy have a syntax for multiple comments.

To add a multiple comment you could insert a # for each line:

**Example**

```python
#This is a comments
#written in
#more than just one line
print("Hello, World!")
```

Or, not quite as intended, you can use a multiple string.

Since Python will ignore string literals that are not assigned to a variable, you can add a multiple string (triple quotes) in your
code, and place your comment inside it:

**Example**

```python
"""
This is a comment
written in
more than just one line
"""
print("Hello, World!")
```

```python
As long as the string is not assigned to a variable, Python will read the code
```
