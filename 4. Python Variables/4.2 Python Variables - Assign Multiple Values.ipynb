## Many Values to Multiple Variables

Python allows hyou to assign values to multiple variables in one line:

## Example

```python
x, y, z = "Orange", "Banana", "Cherry"
print(x)
print(y)
print(z)
```

## One Value to Multiple Variables

And you can assign the same value to multiple variables in one line:

## Example

```python
x = y = z = "Orange"
print(x)
print(y)
print(z)
```

## Unpack a Collection

If you have a collection of values in a list, tuple ect. Python allows you to extract the values into variables. THis is called unpacking.

## Example

Unpack a list:

```python
fruits = ["apple", "banana", "cherry"]
x, y, z = fruits
print(x)
print(y)
print(z)
```
